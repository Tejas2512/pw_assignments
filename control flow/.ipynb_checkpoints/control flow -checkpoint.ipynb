# Basic If-Else Statements:

```python
# 1. Write a Python program to check if a given number is positive or negative.

number = float(input("Enter a number: "))

if number > 0:
    print("The number is positive.")
elif number < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```

```python
# 3. Develop a program to find the maximum of two numbers using if-else statements.

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

if num1 > num2:
    maximum = num1
else:
    maximum = num2

print("The maximum number between", num1, "and", num2, "is", maximum)
```

```python
# 4. Write a Python script to classify a given year as a leap year or not.

year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

```python
# 5. Create a program that checks whether a character is a vowel or a consonant.

character = input("Enter a character: ")

character = character.lower()

if character.isalpha() and len(character) == 1:
    if character in "aeiou":
        print(character, "is a vowel.")
    else:
        print(character, "is a consonant.")
else:
    print("Please enter a single alphabet character.")
```

```python
# 6. Implement a program to determine whether a given number is even or odd.

number = int(input("Enter a number: "))

if number % 2 == 0:
    print(number, "is even.")
else:
    print(number, "is odd.")
```

```python
# 7. Write a Python function to calculate the absolute value of a number without using the `abs()` function.

def absolute_value(number):
    if number < 0:
        return -number
    else:
        return number

number = float(input("Enter a number: "))

result = absolute_value(number)
print("The absolute value of", number, "is", result)
```

```python
# 8. Develop a program that determines the largest of three given numbers using if-else statements.

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))

if num1 >= num2 and num1 >= num3:
    largest = num1
elif num2 >= num1 and num2 >= num3:
    largest = num2
else:
    largest = num3

print("The largest number among", num1, ",", num2, ", and", num3, "is", largest)

```

```python
# 9. Create a program that checks if a given string is a palindrome.

input_string = input("Enter a string: ")

cleaned_string = input_string.replace(" ", "").lower()

if cleaned_string == cleaned_string[::-1]:
    print(input_string, "is a palindrome.")
else:
    print(input_string, "is not a palindrome.")
```

```python
# 10. Write a Python program to calculate the grade based on a student's score.

score = float(input("Enter the student's score: "))

grading_scale = {
    'A': (90, 100),
    'B': (80, 89),
    'C': (70, 79),
    'D': (60, 69),
    'F': (0, 59)
}

grade = None
for letter, (lower_bound, upper_bound) in grading_scale.items():
    if lower_bound <= score <= upper_bound:
        grade = letter
        break

if grade:
    print("The student's grade is", grade)
else:
    print("Invalid score. Please enter a score between 0 and 100.")

```

# Nested If-Else Statements:

```python
# 11. Write a program to find the largest among three numbers using nested if-else statements.

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))

if num1 >= num2:
    if num1 >= num3:
        largest = num1
    else:
        largest = num3
else:
    if num2 >= num3:
        largest = num2
    else:
        largest = num3

print("The largest number among", num1, ",", num2, ", and", num3, "is", largest)
```

```python
# 12. Implement a program to determine if a triangle is equilateral, isosceles, or scalene.

side1 = float(input("Enter the length of the first side: "))
side2 = float(input("Enter the length of the second side: "))
side3 = float(input("Enter the length of the third side: "))

if side1 + side2 > side3 and side1 + side3 > side2 and side2 + side3 > side1:
    if side1 == side2 == side3:
        triangle_type = "equilateral"
    elif side1 == side2 or side1 == side3 or side2 == side3:
        triangle_type = "isosceles"
    else:
        triangle_type = "scalene"

    print("It is a", triangle_type, "triangle.")
else:
    print("It is not a valid triangle.")
```

```python
# 13. Develop a program that checks if a year is a leap year and also if it is a century year.

year = int(input("Enter a year: "))

is_leap_year = False
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    is_leap_year = True

is_century_year = year % 100 == 0

if is_leap_year and is_century_year:
    print(year, "is a leap year and a century year.")
elif is_leap_year:
    print(year, "is a leap year but not a century year.")
else:
    print(year, "is not a leap year.")
```

```python
# 17. Write a Python program to calculate the roots of a quadratic equation.

a = float(input("Enter the coefficient a: "))
b = float(input("Enter the coefficient b: "))
c = float(input("Enter the coefficient c: "))

discriminant = b**2 - 4*a*c

if discriminant > 0:
    root1 = (-b + discriminant**0.5) / (2*a)
    root2 = (-b - discriminant**0.5) / (2*a)
    print("Root 1 =", root1)
    print("Root 2 =", root2)
elif discriminant == 0:
    root = -b / (2*a)
    print("Root =", root)
else:
    real_part = -b / (2*a)
    imaginary_part = (-discriminant)**0.5 / (2*a)
    root1 = complex(real_part, imaginary_part)
    root2 = complex(real_part, -imaginary_part)
    print("Root 1 =", root1)
    print("Root 2 =", root2)
```

```python
# 20. Develop a program that checks if a given number is prime or not using nested if-else statements.

number = int(input("Enter a number: "))

if number > 1:
    is_prime = True
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            is_prime = False
            break
    if is_prime:
        print(number, "is a prime number.")
    else:
        print(number, "is not a prime number.")
else:
    print(number, "is not a prime number.")
```

```python
# 21. Write a Python program to assign grades based on different ranges of scores using elif statements.

score = float(input("Enter the student's score: "))

if 90 <= score <= 100:
    grade = 'A'
elif 80 <= score < 90:
    grade = 'B'
elif 70 <= score < 80:
    grade = 'C'
elif 60 <= score < 70:
    grade = 'D'
else:
    grade = 'F'

print("The student's grade is:", grade)
```

```python
# 22. Implement a program to determine the type of a triangle based on its angles.

angle1 = float(input("Enter the first angle in degrees: "))
angle2 = float(input("Enter the second angle in degrees: "))
angle3 = float(input("Enter the third angle in degrees: "))

if angle1 + angle2 + angle3 == 180:
    if angle1 < 90 and angle2 < 90 and angle3 < 90:
        triangle_type = "acute"
    elif angle1 == 90 or angle2 == 90 or angle3 == 90:
        triangle_type = "right"
    else:
        triangle_type = "obtuse"
    print("The triangle is a", triangle_type, "triangle.")
else:
    print("Invalid input. The angles do not form a triangle.")
```

```python
# 23. Develop a program to categorize a given person's BMI into underweight, normal, overweight, or obese using 
# elif statements.

weight = float(input("Enter the person's weight in kilograms: "))
height = float(input("Enter the person's height in meters: "))

bmi = weight / (height ** 2)

if bmi < 18.5:
    category = "underweight"
elif 18.5 <= bmi < 24.9:
    category = "normal weight"
elif 24.9 <= bmi < 29.9:
    category = "overweight"
else:
    category = "obese"

print("The person's BMI is {:.2f}, which is considered '{}'.".format(bmi, category))
```

```python
# 25. Write a Python script to determine the type of a character (uppercase, lowercase, or special) using elif 
# statements.

character = input("Enter a character: ")

if len(character) == 1:
    if character.isalpha():
        if character.islower():
            char_type = "lowercase"
        else:
            char_type = "uppercase"
    elif character.isdigit():
        char_type = "digit"
    else:
        char_type = "special character"

    print(f"The character '{character}' is a {char_type}.")

else:
    print("Enter single character")
```

```python
# 26. Implement a program to calculate the discounted price based on different purchase amounts using elif 
# statements.

purchase_amount = float(input("Enter the purchase amount: "))

# Calculate the discounted price based on different purchase amounts
if purchase_amount < 50:
    discount = 0
elif purchase_amount < 100:
    discount = 5
elif purchase_amount < 200:
    discount = 10
else:
    discount = 15

discounted_price = purchase_amount - (purchase_amount * discount / 100)

print("The discounted price is: Rs.{:.2f}".format(discounted_price))

```

```python
# 27. Develop a program to calculate the electricity bill based on different consumption slabs using elif 
# statements.

consumption = float(input("Enter the electricity consumption in kWh: "))

rate_slab1 = 0.25  # Rate for the first 100 kWh
rate_slab2 = 0.30  # Rate for the next 200 kWh
rate_slab3 = 0.40  # Rate for consumption over 300 kWh

bill_amount = 0

if consumption <= 100:
    bill_amount = consumption * rate_slab1
elif consumption <= 300:
    bill_amount = 100 * rate_slab1 + (consumption - 100) * rate_slab2
else:
    bill_amount = 100 * rate_slab1 + 200 * rate_slab2 + (consumption - 300) * rate_slab3

print("The electricity bill is: Rs.{:.2f}".format(bill_amount))
```

```python
# 28. Create a program to determine the type of quadrilateral based on its angles and sides using elif 
# statements.

angle1 = float(input("Enter the first angle in degrees: "))
angle2 = float(input("Enter the second angle in degrees: "))
angle3 = float(input("Enter the third angle in degrees: "))
angle4 = float(input("Enter the fourth angle in degrees: "))
side1 = float(input("Enter the length of the first side: "))
side2 = float(input("Enter the length of the second side: "))
side3 = float(input("Enter the length of the third side: "))
side4 = float(input("Enter the length of the fourth side: "))

if angle1 == angle2 == angle3 == angle4 == 90:
    if side1 == side2 == side3 == side4:
        quadrilateral_type = "square"
    else:
        quadrilateral_type = "rectangle"
elif angle1 == angle2 == angle3 == angle4:
    if side1 == side2 == side3 == side4:
        quadrilateral_type = "rhombus"
    else:
        quadrilateral_type = "parallelogram"
else:
    quadrilateral_type = "general quadrilateral"

print("The quadrilateral is a", quadrilateral_type)
```

```python
# 29. Write a Python script to determine the season based on a user-provided month using elif statements.

# (1 for January, 2 for February, etc.)
month = int(input("Enter the month (1-12): "))

if 1 <= month <= 2 or month == 12:
    season = "winter"
elif 3 <= month <= 5:
    season = "spring"
elif 6 <= month <= 8:
    season = "summer"
elif 9 <= month <= 11:
    season = "autumn"
else:
    season = "invalid month"

if season != "invalid month":
    print("The season for month", month, "is", season)
else:
    print("Invalid input. Please enter a valid month (1-12).")

```

```python
# 30. Implement a program to determine the type of a year (leap or common) and month (30 or 31 days) using 
# elif statements

year = int(input("Enter a year: "))

#  (1 for January, 2 for February, etc.)
month = int(input("Enter the month (1-12): "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    year_type = "leap year"
else:
    year_type = "common year"

if 1 <= month <= 12:
    if month in [4, 6, 9, 11]:
        days_in_month = 30
    elif month == 2:
        days_in_month = 29 if year_type == "leap year" else 28
    else:
        days_in_month = 31
else:
    year_type = "invalid month"
    days_in_month = "N/A"

if year_type != "invalid month":
    print("The year", year, "is a", year_type)
    print("The month", month, "has", days_in_month, "days.")
else:
    print("Invalid input. Please enter a valid month (1-12).")
```
