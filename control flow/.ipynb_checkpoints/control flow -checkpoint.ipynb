```python
# 9. Create a program to calculate the total cost of a shopping cart based on discounts.

def calculate_discounted_price(item_price, quantity):
    # Assuming a discount of 10% for a total cost above $100
    discount_threshold = 100
    discount_percentage = 10

    total_cost = item_price * quantity

    # Apply discount if the total cost is above the threshold
    if total_cost > discount_threshold:
        discount_amount = (discount_percentage / 100) * total_cost
        discounted_price = total_cost - discount_amount
        return discounted_price
    else:
        return total_cost

def calculate_total_cost(shopping_cart):
    total_cost = 0

    for item in shopping_cart:
        item_price = item["price"]
        quantity = item["quantity"]
        discounted_price = calculate_discounted_price(item_price, quantity)
        total_cost += discounted_price

    return total_cost

# Example usage:
cart = [
    {"item": "Product 1", "price": 25.0, "quantity": 2},
    {"item": "Product 2", "price": 30.0, "quantity": 1},
    {"item": "Product 3", "price": 20.0, "quantity": 3},
]

total_cost = calculate_total_cost(cart)
print(f"The total cost of the shopping cart is: ${total_cost:.2f}")
```

```python
# 11. Write a program that calculates the roots of a quadratic equation .

import math

def calculate_roots(a, b, c):
    # Calculate the discriminant
    discriminant = b**2 - 4*a*c

    # Check if the discriminant is positive, negative, or zero
    if discriminant > 0:
        # Two distinct real roots
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return root1, root2
    elif discriminant == 0:
        # One real root (repeated)
        root = -b / (2*a)
        return root,
    else:
        # Complex roots
        real_part = -b / (2*a)
        imaginary_part = math.sqrt(abs(discriminant)) / (2*a)
        root1 = complex(real_part, imaginary_part)
        root2 = complex(real_part, -imaginary_part)
        return root1, root2

# Example usage:
a = float(input("Enter the coefficient a: "))
b = float(input("Enter the coefficient b: "))
c = float(input("Enter the coefficient c: "))

roots = calculate_roots(a, b, c)

if len(roots) == 1:
    print(f"There is one real root: {roots[0]:.2f}")
elif len(roots) == 2:
    print(f"There are two real roots: {roots[0]:.2f} and {roots[1]:.2f}")
else:
    print(f"There are complex roots: {roots[0]} and {roots[1]}")
```

```python
# 13. Develop a program that calculates the factorial of a given number using recursion.

def calculate_factorial(n):
    # Base case: factorial of 0 or 1 is 1
    if n == 0 or n == 1:
        return 1
    else:
        # Recursive case: n! = n * (n-1)!
        return n * calculate_factorial(n - 1)

# Example usage:
user_input = int(input("Enter a number to calculate its factorial: "))

# Check if the input is non-negative
if user_input < 0:
    print("Please enter a non-negative integer.")
else:
    factorial_result = calculate_factorial(user_input)
    print(f"The factorial of {user_input} is: {factorial_result}")
```

```python
# 16. Build a program that checks if a given string is a palindrome or not.

def is_palindrome(input_string):
    # Convert the input string to lowercase and remove spaces
    cleaned_string = input_string.lower().replace(" ", "")
    
    # Check if the cleaned string is equal to its reverse
    return cleaned_string == cleaned_string[::-1]

# Example usage:
user_input = input("Enter a string: ")

if is_palindrome(user_input):
    print(f"{user_input} is a palindrome.")
else:
    print(f"{user_input} is not a palindrome.")
```

```python
# 17. Write a program that calculates the average of a list of numbers, excluding the smallest and largest values.

def calculate_average_excluding_extremes(numbers):
    # Check if the list has at least three elements
    if len(numbers) < 3:
        return "Not enough elements to calculate the average."

    # Exclude the smallest and largest values
    numbers.sort()
    trimmed_numbers = numbers[1:-1]

    # Calculate the average of the remaining values
    average = sum(trimmed_numbers) / len(trimmed_numbers)
    return average

# Example usage:
user_input = input("Enter a list of numbers separated by spaces: ")

# Convert the input string to a list of floats
numbers_list = [float(x) for x in user_input.split()]

average_result = calculate_average_excluding_extremes(numbers_list)
print(f"The average (excluding extremes) is: {average_result:.2f}")
```

```python
# 18. Develop a program that converts a given temperature from Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Example usage:
celsius_temperature = float(input("Enter the temperature in Celsius: "))

fahrenheit_temperature = celsius_to_fahrenheit(celsius_temperature)
print(f"{celsius_temperature:.2f} Celsius is equal to {fahrenheit_temperature:.2f} Fahrenheit.")
```

```python
# 19. Create a program that simulates a basic calculator for addition, subtraction, multiplication, and division.

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    # Check if the denominator is zero to avoid division by zero
    if y != 0:
        return x / y
    else:
        return "Error: Division by zero"

# Example usage:
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

print("\nSelect operation:")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

choice = input("Enter choice (1/2/3/4): ")

if choice in ('1', '2', '3', '4'):
    if choice == '1':
        result = add(num1, num2)
        operation = 'Addition'
    elif choice == '2':
        result = subtract(num1, num2)
        operation = 'Subtraction'
    elif choice == '3':
        result = multiply(num1, num2)
        operation = 'Multiplication'
    else:
        result = divide(num1, num2)
        operation = 'Division'

    print(f"\n{operation} result: {result}")
else:
    print("Invalid input. Please enter a valid choice (1/2/3/4).")
```

```python
# 31. Create a program that validates a password based on complexity rules (length, characters, etc.).

def is_valid_password(password, is_digit, is_special_char, is_upper, is_lower, min_length):
    if len(password) < min_length:
        return False, "Password must contain at least {} characters".format(min_length)
    
    if is_upper and not any(char.isupper() for char in password):
        return False, "Password must contain at least one uppercase letter"
    
    if is_lower and not any(char.islower() for char in password):
        return False, "Password must contain at least one lowercase letter"
    
    if is_digit and not any(char.isdigit() for char in password):
        return False, "Password must contain at least one digit"
    
    special_characters = "!@#$%^&*()_-+={}[]|\:;<>,.?/~"
    if is_special_char and not any(char in special_characters for char in password):
        return False, "Password must contain at least one special character"
    
    return True, None

# Input password and required validation checks
password = input("Enter a password: ")

if input("Do you want to countinue with default checks (y/n)?").lower() == "y":  
    is_digit = input("Digit check (y/n)?").lower() == "y"
    is_special_char = input("Special character check (y/n)?").lower() == "y"
    is_upper = input("Upper case check (y/n)?").lower() == "y"
    is_lower = input("Lower case check (y/n)?").lower() == "y"
    min_length = int(input("Min length of password: "))

else:
    is_digit, is_special_char, is_lower, is_upper, min_lenght = True,True,True,True,8
    
# Call the function to validate the password
is_valid, message = is_valid_password(password, is_digit, is_special_char, is_upper, is_lower, min_length)

# Print the result
if is_valid:
    print("Password is valid.")
else:
    print(f"Password is not valid. {message}")
```

```python
# 32. Develop a program that performs matrix addition and subtraction based on user input.

# Input the number of rows and columns
rows = int(input("Enter number of rows:"))
columns = int(input("Enter number of columns:"))

# Input comma-separated values for the first and second matrices
mat1_input = input("Enter a comma-separated list of values for the 1st matrix: ")
mat2_input = input("Enter a comma-separated list of values for the 2nd matrix: ")

# Use list comprehension to convert string values to floating-point numbers
mat1_values = [float(x.strip()) for x in mat1_input.split(',')]
mat2_values = [float(x.strip()) for x in mat2_input.split(',')]

# Function to perform matrix addition or subtraction
def addition_subtraction(matrix1, matrix2, is_add=True):
    result = []
    for row1, row2 in zip(matrix1, matrix2):
        rw = []
        for ele1, ele2 in zip(row1, row2):
            if is_add:
                rw.append(ele1 + ele2)
            else:
                rw.append(ele1 - ele2)
        result.append(rw)
    return result

# Check if the matrices have the correct dimensions
if (len(mat2_values) == len(mat1_values)) and (len(mat1_values) == rows * columns):
    
    # Convert the input values into matrices
    matrix1 = [[mat1_values[i * columns + j] for j in range(columns)] for i in range(rows)]
    matrix2 = [[mat2_values[i * columns + j] for j in range(columns)] for i in range(rows)]

    # Perform matrix addition and subtraction
    addition_result = addition_subtraction(matrix1, matrix2)
    subtraction_result = addition_subtraction(matrix1, matrix2, is_add=False)
    
    # Print the results
    print("\nMatrix Addition Result:")
    for row in addition_result:
        print(row)
    
    print("\nMatrix Subtraction Result:")
    for row in subtraction_result:
        print(row)
        
else:
    print("Invalid matrix dimensions.")
```

```python
# 35. Create a program that simulates a basic text-based tic-tac-toe game against the computer.

import random

# Create a 3x3 matrix filled with False
matrix = [[False, False, False], [False, False, False], [False, False, False]]


def validation(m, turn):
    # Check if any player has won
    # Check rows, columns
    for i in range(3):
        if (m[i][0] == m[i][1] == m[i][2] == turn) or (m[0][i] == m[1][i] == m[2][i] == turn):
            return False, f"{turn} won!"
    
    # Checks diagonals
    if (m[0][0] == m[1][1] == m[2][2] == turn) or (m[0][2] == m[1][1] == m[2][0] == turn):
        return False, f"{turn} won!"

    # Check for a draw
    if all(value is not False for row in m for value in row):
        return False, "It's a draw!"

    return True, ""


def computer_turn(matrix):
    flag = True
    # Generate random coordinates to fill "x" untill found empty cell (False value)
    while flag:
        random_row = random.randint(0, 2)
        random_col = random.randint(0, 2)
        if not matrix[random_row][random_col]:
            flag = False

    matrix[random_row][random_col] = "x"
    return matrix


def user_turn(matrix):
    usr_rw = int(input("row"))
    usr_col = int(input("column"))
    
    # checks given coordinates are empty or not (False value) 
    if not matrix[usr_rw][usr_col]:
        matrix[usr_rw][usr_col] = "o"
        return matrix
    
    # restrict user to update existing value in cell
    else:
        print("Please choose other field, it's already filled!")
        return user_turn(matrix)

# print board after every turn
def updated_matrix(matrix):
    print("\n")
    for row in matrix:
        print(row)


game_on = True
current_turn = "x"

while game_on:
    
    print("\nComputer's turn:")
    matrix = computer_turn(matrix)
    updated_matrix(matrix)
    current_turn = "x"

    game_on, result_message = validation(matrix, current_turn)
    if not game_on:
        print(result_message)
        break

    print("\nYour turn:")
    matrix = user_turn(matrix)
    updated_matrix(matrix)
    current_turn = "o"

    game_on, result_message = validation(matrix, current_turn)
    if not game_on:
        print(result_message)
        break
```
